# Phase 0: Dataset Acquisition

Generate a synthetic dataset for inventory data with 1,000 rows.

## Import Required Libraries

We need pandas for data manipulation and numpy for random number generation.

```python
import pandas as pd
import numpy as np
```

## Generate Synthetic Data

Create 1,000 rows of inventory data with the required columns: QuantityInStock, DaysUntilExpiry, ProductCategory, HistoricalSales.

```python
# Set random seed for reproducibility
np.random.seed(42)

# Number of rows
n_rows = 1000

# Generate QuantityInStock: normal distribution around 500, std 100, positive
quantity = np.random.normal(500, 100, n_rows)
quantity = np.clip(quantity, 0, None)  # Ensure non-negative

# Generate DaysUntilExpiry: exponential distribution, mean 30 days
days_until_expiry = np.random.exponential(30, n_rows)

# Generate ProductCategory: categorical
categories = ['Medicine', 'Equipment', 'Supplies', 'Dairy', 'Produce', 'Bakery']
category_probs = [1/6]*6  # Equal probability for 6 categories
product_category = np.random.choice(categories, n_rows, p=category_probs)

# Simulate HistoricalSales logically
# Base sales + quantity factor + expiry factor + category bonus
base_sales = 100
quantity_factor = 0.1
expiry_factor = 0.5
category_bonuses = {'Medicine': 50, 'Equipment': 30, 'Supplies': 20, 'Dairy': 40, 'Produce': 35, 'Bakery': 45}

historical_sales = []
for i in range(n_rows):
    bonus = category_bonuses[product_category[i]]
    sales = base_sales + quantity_factor * quantity[i] + expiry_factor * days_until_expiry[i] + bonus
    # Add some noise
    sales += np.random.normal(0, 20)
    sales = max(0, sales)  # Ensure non-negative
    historical_sales.append(sales)

# Create DataFrame
df = pd.DataFrame({
    'QuantityInStock': quantity,
    'DaysUntilExpiry': days_until_expiry,
    'ProductCategory': product_category,
    'HistoricalSales': historical_sales
})
```

## Save Dataset to CSV

Save the generated DataFrame to a CSV file for use in subsequent phases.

```python
# Save to CSV
df.to_csv('inventory_data.csv', index=False)

print("Synthetic dataset generated and saved as 'inventory_data.csv'")
print(f"Shape: {df.shape}")
print(df.head())
```
