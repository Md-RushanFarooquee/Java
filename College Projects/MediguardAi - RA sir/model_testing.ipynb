# Phase 1: Interactive Model Testing

Load the trained XGBoost model and predict sales based on user input.

## Import Libraries and Load Model

```python
import pickle
import pandas as pd

# Load the trained model
with open('demand_model.pkl', 'rb') as f:
    model = pickle.load(f)

# Load the encoder
with open('encoder.pkl', 'rb') as f:
    encoder = pickle.load(f)

print("Model and encoder loaded successfully.")
```

## Get User Input

```python
# Set user input values here and run this cell
quantity = 0.0    # current stock
days = 10.0       # days until expiry
category = "Dairy"  # choose Medicine, Equipment, Supplies, Dairy, Produce, or Bakery

print(f"\nInput: Quantity={quantity}, Days={days}, Category={category}")
```

## Preprocess Input and Predict

```python
# Preprocess input
input_df = pd.DataFrame([[category]], columns=['ProductCategory'])
encoded_input = encoder.transform(input_df)
encoded_df = pd.DataFrame(encoded_input, columns=encoder.get_feature_names_out(['ProductCategory']))

# Combine with numerical features
features_df = pd.concat([pd.DataFrame([[quantity, days]], columns=['QuantityInStock', 'DaysUntilExpiry']), encoded_df], axis=1)

# Predict
prediction = model.predict(features_df)[0]

print(f"Predicted Historical Sales: {prediction:.2f}")
```
